```python
#insertion sort

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
#selection sort

def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

```python
#bubble sort

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
#merge sort

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
```

```python
#quick sort

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

```python
#count sort

def counting_sort(arr):
    max_val = max(arr)
    count = [0] * (max_val + 1)

    for num in arr:
        count[num] += 1

    sorted_arr = []
    for i in range(len(count)):
        sorted_arr.extend([i] * count[i])

    return sorted_arr
```

```python
#bucket sort

def bucket_sort(arr):
    if not arr:
        return arr

    max_val = max(arr)
    size = max_val // len(arr) + 1
    buckets = [[] for _ in range(size)]

    for num in arr:
        index = num // len(arr)
        buckets[index].append(num)

    sorted_arr = []
    for bucket in buckets:
        sorted_arr.extend(sorted(bucket))  # Sort each bucket and extend the result

    return sorted_arr
```

```python
#heap sort

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
    return arr
```

```python
#radix sort

def counting_sort_radix(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    for i in range(n - 1, -1, -1):
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1

    for i in range(len(arr)):
        arr[i] = output[i]

def radix_sort(arr):
    max_val = max(arr)
    exp = 1
    while max_val // exp > 0:
        counting_sort_radix(arr, exp)
        exp *= 10
    return arr
```

```python
def plot_multiple_lines(x_values, y_values_dict):
    plt.figure(figsize=(12, 8))
    plt.title('Comparison Line Graph')
    plt.xlabel('Array Size')
    plt.ylabel('Time (MilliSeconds)')
    plt.grid(True)

    for label, y_values in y_values_dict.items():
        plt.plot(x_values, [value*1000 for value in y_values], label=label, marker='o', linewidth=2, alpha=0.9)

    plt.yscale('log')
    plt.legend(loc='best')
    plt.show()

y_values_dict = {
    'Bubble Sort': dict_bubble['Bubble Sort'],
    'Selection Sort': dict_selection['Selection Sort'],
    'Insertion Sort': dict_insertion['Insertion Sort'],
    'Merge Sort': dict_merge['Merge Sort'],
    'Quick Sort': dict_quick['Quick Sort'],
    'Heap Sort': dict_heap['Heap Sort'],
    'Count Sort': dict_count['Count Sort'],
    'Radix Sort': dict_radix['Radix Sort'],
    'Bucket Sort': dict_bucket['Bucket Sort']
}

plot_multiple_lines(x_axis, y_values_dict)
```
